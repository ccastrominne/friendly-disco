```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
plt.style.use ('seaborn-white')
```

```python
def f(x):
    return 2*x
```

```python
def g(x):
    return x**2
```

```python
x = np.linspace(-3,3,1000)
```

```python
plt.figure()
plt.plot(x,f(x), label = '$f(x)$')
plt.plot(x,g(x), label = '$g(x)$')
plt.legend()
#plt.plot(x,f(x),'bo')
```
